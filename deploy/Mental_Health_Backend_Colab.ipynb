```python
import subprocess, sys, os

CRASH_SENTINEL = "/content/.numpy_fixed"

# Check numpy version from a subprocess (avoids loading it in this kernel)
py = sys.executable
ver = subprocess.run([py, "-c", "import numpy; print(numpy.__version__)"], capture_output=True, text=True).stdout.strip()
print("Current numpy version on disk:", ver)

parts = ver.split(".")
major, minor = int(parts[0]), int(parts[1])

if major > 2 or (major == 2 and minor > 4):
    print("numpy too new for scipy. Installing 2.2.0...")
    subprocess.check_call([py, "-m", "pip", "install", "-q", "--force-reinstall", "--no-deps", "numpy==2.2.0"])
    # Verify
    ver2 = subprocess.run([py, "-c", "import numpy; print(numpy.__version__)"], capture_output=True, text=True).stdout.strip()
    print("After install:", ver2)
    # Mark success and crash the kernel so next start uses new numpy
    open(CRASH_SENTINEL, "w").close()
    print("\n--- NUMPY FIXED ---")
    print("Go to Runtime > Restart runtime, then Runtime > Run all")
    os._exit(0)
else:
    print("numpy version OK")
    # Remove sentinel if present from previous crash
    if os.path.exists(CRASH_SENTINEL):
        os.unlink(CRASH_SENTINEL)
```

```python
import threading, time

def run_flask():
    import app
    app.app.run(host='0.0.0.0', port=5000, debug=False, use_reloader=False)

print("Starting Flask server on port 5000...")
flask_thread = threading.Thread(target=run_flask, daemon=True)
flask_thread.start()
time.sleep(2)
print("Flask server is running.")
```

```python
import re, subprocess, sys

print("Starting Cloudflare tunnel...")
proc = subprocess.Popen(
    ['/content/cloudflared', 'tunnel', '--url', 'http://localhost:5000'],
    stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True
)

url = None
for line in proc.stderr:
    sys.stdout.flush()
    m = re.search(r'https://[a-zA-Z0-9-]+\.trycloudflare\.com', line)
    if m:
        url = m.group()
        print(f"\n{'=' * 60}")
        print(f"  BACKEND PUBLIC URL")
        print(f"  {url}")
        print(f"{'=' * 60}")
        print(f"\nCopy this URL. Then in vercel.json change:")
        print(f'  "source": "/api/(.*)",')
        print(f'  "destination": "{url}/$1"')
        print(f"\nCommit and push to Vercel. The frontend will reach your Colab backend.")
        break
```
