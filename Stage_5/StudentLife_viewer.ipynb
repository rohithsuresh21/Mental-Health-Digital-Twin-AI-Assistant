```python
activity = pyreadr.read_r(
    f"{BASE}/sensing/activity.Rds"
)[None]

print(activity.shape)
print(activity.columns)
activity.head()
```

```python
import pandas as pd

activity["datetime"] = pd.to_datetime(
    activity["timestamp"],
    unit="s"
)

activity["date"] = activity["datetime"].dt.date

activity.head()
```

```python
daily_activity = (
    activity
    .groupby(
        ["uid", "date", "activity_inference"]
    )
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

daily_activity.head()
```
